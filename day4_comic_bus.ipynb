```python
import requests
from pyquery import PyQuery as pq
```

發現有亂碼的時候，就要確認編碼是否正確，可以看 html 查找

```python
r = requests.get('https://comicbus.com/')
r.encoding = 'big5'
```

```python
r.text
```

# 查找每日 hotcomicpic_dl

```python
page = pq(r.text)
daily_block = page("#hotcomicpic_dl")
print(daily_block.html())
```

```python
for block in daily_block(".picborder"):
    print("=================")
    link = pq(block).find("a").attr('href')
    title = pq(block).find("a").attr('title')
    num = pq(block).find("num").text()
    print(f"link = https://comicbus.com/{link}")
    print(f"title = {title}")
    print(f"num = {num}")
```
