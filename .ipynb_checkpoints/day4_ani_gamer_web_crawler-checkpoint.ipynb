# 看來他可能是需要透過 javascript 進行渲染 - 使用這兩年新出的 request_html 套件
requests 的團隊新出的套件，以往 requests 是不支援 javascript 進行渲染的，因此在早年需要透過一些模擬瀏覽器的套件，像是 selenium，現在就不需要了。

Doc: https://docs.python-requests.org/projects/requests-html/en/latest/
- 支持JavaScript
- 支持CSS選擇器（又名jQuery風格, 感謝PyQuery）
- 支持Xpath選擇器
- 可自定義模擬User-Agent（模擬得更像真正的web瀏覽器）
- 自動追蹤重定向
- 連接池與cookie持久化
- 支持異步請求


```python
!pip install requests-html
```

```python
from requests_html import HTMLSession

# 獲取請求對象 
session = HTMLSession() 
# 往巴哈姆特動畫瘋發送get請求 
ani_gamer = session.get('https://ani.gamer.com.tw/') 

ani_gamer.text
```

# 傳統做法 - 利用 Pyquery 解析 html 資料

```python
page = pq(ani_gamer.text)
```

```python
page.html()
```

```python
for anime_name in page.find(".anime-name"):
    print(pq(anime_name))
```

```python
type(page.find(".anime-name"))
```

```python
page.find(".anime-name").eq(0).html()
```

```python
page.find(".anime-name").eq(1).html()
```

```python
page.find(".anime-name").eq(1).text()
```

# 利用 request_html 解析 html 資料

```python
# 獲取請求對象 
session = HTMLSession() 
# 往巴哈姆特動畫瘋發送get請求 
ani_gamer = session.get('https://ani.gamer.com.tw/') 
ani_gamer.text
```

```python
for anime_name in ani_gamer.html.find(".anime-name"):
    anime_name.encoding = 'utf-8'
    print(anime_name.html)
```
