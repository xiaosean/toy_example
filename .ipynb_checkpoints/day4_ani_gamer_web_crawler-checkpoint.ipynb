# 查找每日 hotcomicpic_dl

```python
page = pq(r.text)
daily_block = page("#hotcomicpic_dl")
print(daily_block.html())
```

```python
for block in daily_block(".picborder"):
    print("=================")
    link = pq(block).find("a").attr('href')
    title = pq(block).find("a").attr('title')
    num = pq(block).find("num").text()
    print(f"link = https://comicbus.com/{link}")
    print(f"title = {title}")
    print(f"num = {num}")
```

# 但有些網站利用 request 是無法解析出整個 HTML 的，可能是他有做驗證機制，需要 user_agent 或是 cookie, 另一個可能是他使用 javascript 進行渲染。

```python
r = requests.get('https://seiga.nicovideo.jp/')
r.encoding = 'utf-8'
r.text
```

# 如果無法正常存取 第一步就是塞 Header or cookie 讓他覺得你是瀏覽器

```python
user_agent = "Mozilla/5.0"
headers = {
    "user_agent": user_agent,
#     "cookie": cookies
}
r = requests.get('https://seiga.nicovideo.jp/', headers=headers)
r.encoding = 'utf-8'
r.text
```

# 看來他可能是需要透過 javascript 進行渲染 - 使用這兩年新出的 request_html 套件
requests 的團隊新出的套件，以往 requests 是不支援 javascript 進行渲染的，因此在早年需要透過一些模擬瀏覽器的套件，像是 selenium，現在就不需要了。

Doc: https://docs.python-requests.org/projects/requests-html/en/latest/
- 支持JavaScript
- 支持CSS選擇器（又名jQuery風格, 感謝PyQuery）
- 支持Xpath選擇器
- 可自定義模擬User-Agent（模擬得更像真正的web瀏覽器）
- 自動追蹤重定向
- 連接池與cookie持久化
- 支持異步請求


```python
!pip install requests-html
```

# 常用指令
- eq(index)  # 由於一個文件可能會有多個相同的 class 名稱，因此搜尋回來通常是陣列形式。eq 可看作為 index，從 0 開始
- html() # 會顯現出 html 區塊，通常是用於 Debug
- text() # 從 html 區塊中直接萃取出文字

```python
page.find(".anime-name").eq(0).html()
```

```python
page.find(".anime-name").eq(1).html()
```

```python
page.find(".anime-name").eq(1).text()
```

# Quiz：萃取出這季新番的名字 集數 以及網址連結

```python
for block in page.find(".anime-block"):
    anime_name = pq(block).find(".anime-name_for-marquee").text()
    href =  "https://ani.gamer.com.tw/" + pq(block).find(".anime-card-block").attr("href")
    anime_episode = pq(block).find(".anime-episode").text()
    print(f"======={anime_name}========")
    print(f"最新連載 = {anime_episode}")
    print(f"連結 = {href}")
    print()
```
