# Pandas 可輕鬆創建欄位，我們要創建一個 Feature 是年齡劃分的，推估可能小孩和老年人死亡率會較高

```python
df.loc[ df['Age'] <= 16, 'Mapping_Age'] = 0
df.loc[(df['Age'] > 16) & (df['Age'] <= 32), 'Mapping_Age'] = 1
df.loc[(df['Age'] > 32) & (df['Age'] <= 48), 'Mapping_Age'] = 2
df.loc[(df['Age'] > 48) & (df['Age'] <= 64), 'Mapping_Age'] = 3
df.loc[ df['Age'] > 64, 'Mapping_Age'] = 4 ;
```

```python
age_avg = df["Age"].mean()
df.loc[df["Age"].isnull(), "Age"] = int(age_avg)
```

## Pclass (Passenger Class) is the socio-economic status of the passenger and it is a categorical ordinal feature which has 3 unique values (1, 2 or 3):
1 = Upper Class
2 = Middle Class
3 = Lower Class

```python
df.head()
```

```python
# Mapping titles
pclass_mapping = {1: "Upper Class", 2: "Middle Class", 3: "Lower Class"}
df['Pclass'] = df['Pclass'].map(pclass_mapping)
df['Pclass'] = df['Pclass'].fillna("不知道欸 因為有缺值")
```
