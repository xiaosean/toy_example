```python
momo_page = pq(r.text)
```

```python
data = []
for item in momo_page(".eachGood"):
    special_title = pq(item).find(".specialTitle").text()
    product = pq(item).find(".prdName").text()
    price = pq(item).find(".prdPrice").text()
    data += [[product, price, special_title]]
print(data)
```

```python
df = pd.DataFrame(columns=["product", "price", "special_title"], data=data)
```
