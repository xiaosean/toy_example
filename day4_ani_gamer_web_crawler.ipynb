# 以巴哈姆特動畫瘋作為示範，透過 Pyquery 套件讀取

https://ani.gamer.com.tw/

```python
# 獲取請求對象 
session = HTMLSession() 
# 往巴哈姆特動畫瘋發送get請求 
ani_gamer = session.get('https://ani.gamer.com.tw/') 

ani_gamer.text
```

```python
page = pq(ani_gamer.text)
```

```python
page.html()
```

# 獲得動漫列表的名稱

```python
for anime_name in page.find(".anime-name_for-marquee"):
#     print(pq(anime_name))
    print(pq(anime_name).text())
```

# 常用指令
- eq(index)  # 由於一個文件可能會有多個相同的 class 名稱，因此搜尋回來通常是陣列形式。eq 可看作為 index，從 0 開始
- html() # 會顯現出 html 區塊，通常是用於 Debug
- text() # 從 html 區塊中直接萃取出文字

```python
page.find(".anime-name").eq(0).html()
```

```python
page.find(".anime-name").eq(1).html()
```

```python
page.find(".anime-name").eq(1).text()
```

# Quiz：萃取出這季新番的名字 集數 以及網址連結

```python
for block in page.find(".anime-block"):
    anime_name = pq(block).find(".anime-name_for-marquee").text()
    href =  "https://ani.gamer.com.tw/" + pq(block).find(".anime-card-block").attr("href")
    anime_episode = pq(block).find(".anime-episode").text()
    print(f"======={anime_name}========")
    print(f"最新連載 = {anime_episode}")
    print(f"連結 = {href}")
    print()
```
